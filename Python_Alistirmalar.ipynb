```python
#Görev 5 : Argman olarak bir liste alan listenin içerisinde tek ve çift sayıları ayrı listelere
#atayan ve bu listeleri return eden bir fonksiyon yaz

l = [2, 13, 18, 93, 22]
even_list=[]
odd_list = []

def func():

    for i in l:

     if i % 2 == 0 :
      even_list.append(i)
     elif i % 2 != 0 :
      odd_list.append(i)

    return odd_list,even_list
func()
```

```python
#Görev 6
#Aşağıda veirlen listeden mühendislik ve tıp fakültelerinde dereceye giren öğrencilerin
#isimleri bulunmaktadır. Sırasıyla ilk üç öğrenci mühendislik fakültesinin başarı sırasını
#temsil ederken son üç öğrencide tıp fakültesi öğrenci sırasına aittir. Enumarate
#kullanrak öğrenci derecelerini fakülte özelinde yazınız

ogrenciler = ["Ali","Veli","Ayşe","Talat","Zeynep","Ece"]
```

```python
for i,ogrenci in enumerate(ogrenciler,start=1):

  if i < 4 :
    print(f"Mühendislik Fakültesi {i} . öğrenci : {ogrenci}")
  elif i >= 4 :
    print(f"Tıp Fakültesi {i-3} . öğrenci : {ogrenci}")

```

```python
#görev 7 : Aşağıda 3 liste verilmiştir. Listelerden sırası ile dersin kodu , kredisi
#kontenjan bilgileir yer almaktadır . Zip kullanrak ders bilgilerini bastırın

ders =["CMP1005","PSY1001","HUK1005","SEN2204"]
kredi = [3, 4, 2, 4]
kontenjan = [30, 75, 150, 25]

for x,y,z in zip(ders,kredi,kontenjan):

  print(f"Kredisi {y} olan {x} kodlu dersin kontenjanı {z} kişidir")
```

```python
#Görev 8: aşağıda 2 adet set verilmiştir. Sizden istenilen eğer 1. küme 2. kümayi
#kapsıyor ise ortak elemanlarını eğer kapsamıyor ise 2. kümenin 1. kümeden farkını
#yazdıracak fonksiyonu tanımlamanız bekleniyor.

kume1 = set(["data", "python"])
kume2 = set(["data","function","qcut","lambda","python","miuul"])

def kume_durum(set1,set2):
  if set1.issuperset(set2):
    print(set1.isdisjoint(set2))
  else:
    print(set2.difference(set1))

kume_durum(kume1,kume2)
```
