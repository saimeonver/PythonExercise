```python
def calculate (varm, moisture, charge):
  varm = varm * 2
  moisture = moisture * 2
  charge = charge * 2
  output = (varm + moisture) / charge

  return varm, moisture, charge, output
```

```python
calculate(52,36,10)
```

```python
type(calculate(52,36,10))
```

```python
#fonkisyonumda dönen değerlerin atamasını gerçekleştirdim
varm, moisture, charge, output = calculate(52,36,10)
```

```python
# Fonksiyon içerisinden fonkisyon çağırmak
def calculate(varm, moisture, charge):
 return int((varm + moisture)/charge)
```

```python
calculate(52,36,10)
```

```python
def standardization(a, p):
  return a * 10 / 100 * p * p
```

```python
standardization(56 , 3)
```

```python
#yukarıda yazdığım iki fonksiyonu birleştiricem

def all_calculation(varm, moisture, charge, p):
  a = calculate(varm, moisture, charge)
  b = standardization(a, p)
  print(b*10)
```

```python
all_calculation(25,36,20,6)
```

```python
#Local ve Global Değişkenler (Loca ve Global Variables)

#list store ise global değişendir her yerden erişim olur
list_store = [1, 2]

def add_element(a, b):
  #c local etki alannda
  c = a * b
  list_store.append(c)
  print(list_store)

add_element(5,9)
```
