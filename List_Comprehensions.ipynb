```python
#Comprehensions

#List Comprehension

#Eski Yol
salaries = [1000, 2000, 3000, 4000, 5000]

def new_salary (salary):
  return salary * 20 / 100 + salary

for salary in salaries:

  print(new_salary(salary))
```

```python
#Zam lı hallerini farklı bir listeye ekleyeyim
null_list = []

for salary in salaries:
  null_list.append(new_salary(salary))
```

```python
null_list
```

```python
#Maaş 3000 den fazlayasa %20, azsa farklı işlem yapıcam
null_list = []

for salary in salaries :
  if salary > 3000:
    null_list.append(new_salary(salary))
  else:
    null_list.append(new_salary(salary*2))
```

```python
null_list
```

```python
#Yeni yol

[new_salary(salary*2) if salary < 3000 else new_salary (salary) for salary in salaries]
```

```python
#Maaşları 2 ile çarpıcam

salaries = [1000, 2000, 3000, 4000, 5000]

[salary * 2 for salary in salaries]
```

```python
#Maaşı 3000 den az olnalrı 2 ile çarpalım diğerleri aynen kalacak

[salary * 2 if salary < 3000 else salary for salary in salaries]
```

```python
#Maaşı 3000 den az olnalrı 2 ile çarpalım diğerlerine bir işlem yapılmayacak
[salary *2 for salary in salaries if salary < 3000]
```

```python
#Önemli
#eğer sadece if varsa if for döngüsünün sağına yazılır
#eğer if ile birlikte else varsa if solda else ifin sağında yazılır. for en sağda kalır
```

```python
#fonksiyonda kullanabilirm

[new_salary(salary * 2 )if salary < 3000 else new_salary(salary * 1.5)for salary in salaries]
```

```python
#Tüm öğrenci listemde no olanları küçük no olmayanları büyük har file yazıcam

student = {"John", "Mark", "Venessa", "Mariam"}
students_no = ["John", "Venessa"]

[s.lower() if s in students_no else s.upper() for s in student]
```

```python
[s.upper() if s not in students_no else s.lower() for s in student]
```
