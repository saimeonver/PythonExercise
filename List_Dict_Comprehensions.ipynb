```python
#Dict Comprehension

dictionary = {'a' : 1,
              'b' : 2,
              'c' : 3,
              'd' : 4}
dictionary. keys()
dictionary.values()
dictionary.items()
```

```python
#value değerlerinin karesini alıcam

{k: v ** 2 for (k,v) in dictionary.items()}
```

```python
{k.upper(): v for k,v in dictionary.items()}
```

```python
{k.upper():v**2 for k,v in dictionary.items()}
```

```python
#Uygulama Mülakat Sorusu

#Amaç : çift sayıların karesi alınarak bir sözlüğe eklenmelidir
#Key ler orjinal değerler , value değerleri değiştirilmiş olacak

numbers = range(10)
new_dict = {}

for n in numbers :
  if n % 2 == 0:
    new_dict[n] = n **2

```

```python
new_dict
```

```python
#2. yol
{k : v **2 for k,v in enumerate(range(10)) if k % 2 == 0}
```

```python
#3. yol
{k : k **2  for k in range(10) if k % 2 == 0}
```

```python
#Soru : İsminde "INS" olan değişkenlerin başına FLAG diğerlerine NO_FLAG eklemek istiyorum

df.columns = ["FLAG_" + col if 'INS' in col else "NO_FLAG_" + col for col in df.columns]
```

```python
# Soru : Amaç key i string, valuesu aşağıdaki gibi bir liste olan sözlük oluşturmak
#sadece sayısal değişkenlerde işlem yapıcam

{col : ["mean","min","max","var"] col.items() for col in df.columns}
```

```python
import seaborn as sns
df = sns.load_dataset("car_crashes")
df.columns
```

```python
num_cols = [col for col in df.columns if df[col].dtype != "O"]
soz = {}
agg_list = ["mean","min","max","var"]
```

```python
num_cols
```

```python
for col in num_cols:
  soz[col] = agg_list

soz
```

```python
#Diğer yol

import seaborn as sns
df = sns.load_dataset("car_crashes")
df.columns
```

```python
#İkinci yol
new_dict = {col : agg_list for col in num_cols}
```
