```python
#Set

"""
Değiştirilebilir
Sırası + Eşsizdir
Kapsayıcıdır

"""

#difference(): iki kümenin farkı

set1 = set([1, 3, 5])
set2 = set([1, 2, 3])
```

```python
type(set1)
```

```python
#set1 de olup set2 de olmayanlar
set1.difference(set2)
```

```python
#set2 de olup set1 de olmayanlar
set2.difference(set1)
```

```python
# symmetric_difference(): iki kümede de birbirlerine göre olmayanlar
set1.symmetric_difference(set2)
```

```python
set2.symmetric_difference(set1)
```

```python
# intersection(): iki kümenin kesişimi

set1 = set([1, 3, 5])
set2 = set([1, 2, 3])
```

```python
set1.intersection(set2)
```

```python
set2.intersection(set1)
```

```python
#iki kümenin kesişimi sembolik gösterim
set1 & set2
```

```python
set1 - set2
```

```python
set2 - set1
```

```python
# union (): iki kümenin birleşimi

set1.union(set2)
```

```python
set2.union(set1)
```

```python
#isdisjoint() : iki kümenin kesişimi boş mu?
#is ile başlayınca true false döner

set1 = set([7, 8, 9])
set2= set([5, 6, 7, 8, 9, 10])
```

```python
set1.isdisjoint(set2)
```

```python
set2.isdisjoint(set1)
```

```python
#isubset() : bir küme diğer kümenin alt kümesi mi?

set1.issubset(set2)
```

```python
set2.issubset(set1)
```

```python
#issuperset(): bir küme diğer kümeyi kapsıyor mu?
set2.issuperset(set1)
```

```python
set1.issuperset(set2)
```
